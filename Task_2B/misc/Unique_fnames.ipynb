```python
import os

def find_unique_filenames(directory):
    # Create an empty set to store unique filenames without suffix
    unique_filenames = set()

    # Walk through the directory
    for root, _, files in os.walk(directory):
        for file in files:
            # Split the filename at the first dot
            filename = file.split('.')[0]
            # Add the filename to the set (ignoring the extension)
            unique_filenames.add(filename)

    return unique_filenames

# Example usage
directory_path = r"C:\Users\qiangy\Box\SEACAR_OEAT_FY22-23\SEACAR_WQ_Analysis_Pilot\Working_Folders\data_transfer\SEACAR_WQ_2024\GIS_Data\raster_output\IDW_Month"  # Replace with your directory path
unique_filenames = find_unique_filenames(directory_path)

len(unique_filenames)
```
