# Analysis of RMSE and ME (new data)

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import numpy as np
```

### Read data table

```python
result_folder = r"E:/Projects/SEACAR_WQ_2024/result/result_v3/"
fn = result_folder+ 'rk_idw_comp.csv'  # Replace 'yourfile.csv' with the path to your CSV file
df = pd.read_csv(fn).drop(columns=['Unnamed: 0.1','Unnamed: 0'])
```

### Determine the lowest RMSE and absolute ME

```python
# Compare RMSE between IDW and RK where only IDW can run
df['lowest_RMSE'] = np.where(
    pd.isna(df['RMSE_idw_all']) & pd.isna(df['RMSE_rk']), np.nan, # If both A and B are NaN, assign NaN
    np.where(
        pd.isna(df['RMSE_idw_all']), 'RK',  # If A is NaN but B is not, assign 'y'
        np.where(
            pd.isna(df['RMSE_rk']), 'IDW',  # If B is NaN but A is not, assign 'x'
            np.where(
                df['RMSE_idw_all'] > df['RMSE_rk'], 'RK', 'IDW'  # If A > B, assign 'x', otherwise assign 'y'
            )
        )
    )
)

# Compare abs ME between IDW and RK in all cases
df['lowest_abs_ME'] = np.where(
    pd.isna(df['ME_idw_all']) & pd.isna(df['ME_rk']), np.nan, # If both A and B are NaN, assign NaN
    np.where(
        pd.isna(df['ME_idw_all']), 'RK',  # If A is NaN but B is not, assign 'y'
        np.where(
            pd.isna(df['ME_rk']), 'IDW',  # If B is NaN but A is not, assign 'x'
            np.where(
                df['ME_idw_all'] > df['ME_rk'], 'RK', 'IDW'  # If A > B, assign 'x', otherwise assign 'y'
            )
        )
    )
)

# Identify seasons interpolated by RK and IDW respectively
df.loc[pd.notna(df['RMSE_rk']), 'processed_by_RK'] = 'RK'
df.loc[pd.notna(df['RMSE_idw_all']), 'processed_by_IDW'] = 'IDW'

df1 = df.dropna(subset=['RMSE_idw_all','RMSE_rk'], how='all')
df2 = df.dropna(subset=['RMSE_idw_all','RMSE_rk'])
```

## 4. Distribution of RMSE and ME

#### Define a function to draw the boxplots

```python
def box_plot(df, err_metric_col, err_metric, method, para, fz1, fz2, ax):

    df_rk = df[(df[err_metric_col]>-1000)*(df['Parameter_acr'] == para)].dropna(subset=['processed_by_'+method,err_metric_col])

    boxprops = dict(linestyle='-', linewidth=2, color='k')  # Customizing the box border
    meanprops = dict(marker='D', markeredgecolor='black', markerfacecolor='firebrick')  # Customizing the mean marker
    medianprops = dict(linestyle='-', linewidth=2, color='black')  # Customizing the median line color

    df_rk.boxplot(by='WaterBody_acr', column=[err_metric_col],boxprops=boxprops, showmeans=True, meanprops=meanprops, meanline=False, medianprops=medianprops, patch_artist=True, ax = ax)  # Specify the column for which you want the box plot
    ax.set_title(para,fontsize=fz1)  # Optional: Adds a title to the plot
    ax.set_ylabel(err_metric,fontsize=fz2)  # Optional: Label the y-axis
    ax.set_xlabel('Water bodies',fontsize=fz2)  # Optional: Label the y-axis
```

### 4.1 Distribution of RMSE by RK

```python
fig, axs = plt.subplots(2, 3, figsize=(15, 10))
axs = axs.flatten()

err_metric_col, err_metric, method = 'RMSE_rk', 'RMSE', 'RK'
n=0
for para in df['Parameter_acr'].unique():
    box_plot(df, err_metric_col, err_metric, method, para, 16, 14, axs[n])
    n = n+1
    

# Set main title with adjusted vertical position
fig.suptitle('{} of {} interpolation'.format(err_metric,method), fontsize=20, y=1)

# Optionally adjust top margin
fig.tight_layout()

# Show plot
plt.show()
```

## 5. Relation between accuracy and sample size

### The function to create scatter plots and regression analysis

```python
df
```

```python
def scatter_plot(water_body, err_metric, method, err_metric_col,fz1, fz2, ax):
    #df_nn = df.dropna(subset=['RMSE_idw_all'])
    df_nn = df[(df['WaterBody'] == water_body) & (df[err_metric_col] > -10000)].dropna(subset=[err_metric_col])

    # Calculate abs ME
    df_nn[err_metric_col] = df_nn[err_metric_col].abs()

    # Regression analysis
    # Add a constant term for the intercept
    X = sm.add_constant(df_nn['NumDataPoints_all'])
    model = sm.OLS(df_nn[err_metric_col], X).fit()

    # Fit a linear trendline
    slope, intercept = np.polyfit(df_nn['NumDataPoints_all'], df_nn[err_metric_col], 1)
    p_value = model.pvalues[1]

    # Creating a scatter plot using pandas
    df_nn.plot(kind='scatter', x='NumDataPoints_all', y=err_metric_col, ax = ax)

    # Add the trendline
    ax.plot(df_nn['NumDataPoints_all'], slope*df_nn['NumDataPoints_all'] + intercept, color='red', label=f'Trendline: y={slope:.2e}x+{intercept:.3f}')


    ax.set_title('{} of {} in {}'.format(err_metric,method,water_body),fontsize=fz1)
    ax.set_xlabel('Sample size',fontsize=fz2)
    ax.set_ylabel(err_metric,fontsize=fz2)

    # Add a constant term for the intercept
    X = sm.add_constant(df_nn['NumDataPoints_all'])
    model = sm.OLS(df_nn[err_metric_col], X).fit()

    # Step 4: Add the regression equation to the plot
    equation_and_pvalue = f'y = {slope:.3f}x + {intercept:.3f} (p={p_value:.3g})'
    ax.text(min(df_nn['NumDataPoints_all']), max(df_nn[err_metric_col])*0.95, equation_and_pvalue, fontsize=fz2, verticalalignment='bottom', color='red')
```

### 5.1 IDW
#### RMSE v.s. Sample points (Charlotte Harbor, Big Bend, Biscayne Bay)

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 4))
scatter_plot('Charlotte Harbor', 'RMSE','IDW','RMSE_idw_all',16,14, axs[0])
scatter_plot('Big Bend Seagrasses', 'RMSE','IDW','RMSE_idw_all',16,14, axs[1])
scatter_plot('Biscayne Bay', 'RMSE','IDW','RMSE_idw_all',16,14, axs[2])
```

#### Abs. ME v.s. Sample points (Charlotte Harbor, Big Bend, Biscayne Bay)

```python
fig, axs = plt.subplots(1, 3, figsize=(16, 4))

scatter_plot('Charlotte Harbor', 'Abs.ME','IDW','ME_idw_all',15,14, axs[0])
scatter_plot('Big Bend Seagrasses', 'Abs.ME','IDW','ME_idw_all',15,14, axs[1])
scatter_plot('Biscayne Bay', 'Abs.ME','IDW','ME_idw_all',15,14, axs[2])
```
